```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from random import randint, seed
from pydataset import data
import seaborn as sns
import env
seed(123)

iris = sns.load_dataset('iris')
# iris = data('iris')
iris
```

```python
# 1. What does the distribution of petal lengths look like?
# Continuous: sepal.Length, Sepal.Width, Petal.Length, Petal.Width
# Categorial: Species
sns.displot(data=iris, x='petal_length');
```

```python
# 2. Is there a relationship between petal length and petal width?
sns.relplot(data=iris, x='petal_length', y='petal_width');
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width');
```

```python
# 3. Would it be reasonable to predict species based on sepal width and sepal length? 
# For this, you'll visualize two numeric columns through the lense of a categorical column.
sns.relplot(data=iris, x='sepal_width', y='sepal_length', hue='species');
```

```python
# 4. Which features would be best used to predict species?
plt.figure(figsize=(8,8))

sns.pairplot(data=iris, corner=True, hue='species')
plt.show()
```
